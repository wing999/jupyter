# 模拟器检测测试

**用于检测是否为模拟器数据。数据来源为官服实际数据**

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv('./data/Simulator.csv')
```

```python
data_onehot_df = pd.get_dummies(data_df,columns=['cpu_abi','host','hardware','device','bootloader','cpu_abi2','os_name','board','manufacturer','build_id_prop','run_mode'])
```

```python
data_onehot_df
```

```python
data = data_onehot_df.drop(['zisemulator'],inplace=False,axis=1)
target = data_onehot_df['zisemulator']
x_train,x_test,y_train,y_test = train_test_split(data,target,random_state=1)
```

```python
mode = LogisticRegression()
mode.fit(x_train,y_train)
mode.score(x_train,y_train)
```

```python
# mode.predict(x_test,y_test)
mode.score(x_test,y_test)
```
