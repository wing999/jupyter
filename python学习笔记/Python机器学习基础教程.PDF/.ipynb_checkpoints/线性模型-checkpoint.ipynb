# 线性模型

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
```

## 线性回归(又名 普通最小二乘法)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
```

```python
X, y = mglearn.datasets.make_wave(n_samples=60)
df = pd.DataFrame(X,columns=['x'])
df['y']=y
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = LinearRegression().fit(X_train, y_train)  #权重系数使用lr.coef_ 获得 截距使用lr.intercept_ 获得
```

```python
print("训练集正确率：",lr.score(X_train,y_train)) # 训练集正确率为67%，所以因为只有一个特征维度，导致出现欠拟合情况
print("测试集正确率：", lr.score(X_test,y_test))
```

### 波士顿房价数据

**波士顿房价数据的特征比较多，可以减少欠拟合的情况，但可能会增加过拟合几率**

```python
from sklearn.datasets import load_boston
```

```python
x,y = mglearn.datasets.load_extended_boston() #得到经过特征工程后的波士顿房价数据
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
x.shape # 经过特征工程后的波士顿房价数据特征数为104
```

```python
lr = LinearRegression().fit(x_train,y_train)
```

```python
lr.score(x_train,y_train) # 训练集的准确率为95% 比之前67%有所提高，但还是存在欠拟合情况
```
