# 常用的概率分布函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 均匀分布[Uniform Distribution]

```python
data = np.random.randint(1,7,10000) #掷骰子 1～6 一万次
```

```python
plt.hist(data,6) #6个柱子
plt.show()  #概率是均匀的，骰子每个面朝上的几率是均等的
```

## 正态分布[Normal / Gaussian Distribution]
概率密度函数是用来描绘连续型随机变量

```python
from scipy.stats import norm # 导入正态分布函数
```

```python
data2 = np.linspace(1.5,1.9,10000) # 1.5~1.9米的身高；生成均匀间隔的10000个数字
```

```python
plt.plot(data2,norm.pdf(data2,loc=1.7,scale=0.05)) # pdf 概率密度函数缩写；将身高变成正太分布
# loc：中间值  scale：标准差，标准差是用来描绘数据的离散程度。【调节中间值和标准差来观察图的变化】
# 平均1.7米，正负5厘米，即1.65～1.75米范围内的身高占大部分

# plot(x,y) 以原数据为x，norm.pdf()求出来的数据为y绘制图形
```

## 二项分布[Binomial Probability Mass Function]
概率质量函数是针对离散型随机变量

```python
from scipy.stats import binom
```

```python
n,p = 10,0.5 # n->抛10个硬币，要么正面朝上，要么反面朝上，成功的概率[正面朝上]为0.5
data3 = np.arange(1,11) # 期望得到1～10次正面朝上
plt.plot(data3,binom.pmf(data3,n,p),'bo')  # pmf针对离散数据，用‘bo’绘制蓝点
# 抛10个硬币，恰好有5个正面朝上的概率为0.25, 10个都为正面朝上几乎不可能
```

## 泊松分布[Poisson Probability Mass Function]
经常用于流量，单位时间内的某种情况的概率。客流量，车流量等

```python
from scipy.stats import poisson
```

```python
mu = 20000 # 假设上海迪士尼 非假期 日均客流量2万人次
data4 = np.arange(19000,21000)  #出现一万九～两万一客流量的概率
plt.plot(data4,poisson.pmf(data4,mu),'bo')  # pmf 针对离散数据，用'bo'绘制蓝点
```

## 指数分布[Exponential Probability Density Function]

```python
from scipy.stats import expon
```

```python
data5 = np.linspace(0,7,10000) # 双11期间 当天～第七天
plt.plot(data5,expon.pdf(data5)*100) # pdf 针对连续型随机变量。每天成交量1~100亿 
```
