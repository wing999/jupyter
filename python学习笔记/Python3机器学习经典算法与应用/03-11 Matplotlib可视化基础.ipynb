# matplotlib 基础

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10 ,100)
x
```

```python
y=np.sin(x)
```

```python
y
```

## 折线图

**画单线图**

```python
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

**画双线图**

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

**修改颜色**

```python
plt.plot(x, siny)
plt.plot(x, cosy, c='r')
plt.show()
```

**修改线的样式**

```python
plt.plot(x, siny)
plt.plot(x, cosy, c='r', linestyle='--')
plt.show()
```

**设置X轴与Y轴坐标范围**

```python
plt.plot(x, siny)
plt.plot(x, cosy, c='r', linestyle='--')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

**设置X轴Y轴坐标范围2**

```python
plt.plot(x, siny)
plt.plot(x, cosy, c='r', linestyle='--')
plt.axis([-1, 11, -2, 2])
plt.show()
```

**设置X轴Y轴含义**

```python
plt.plot(x, siny)
plt.plot(x, cosy, c='r', linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

**设置图例**

```python
plt.plot(x, siny, label='sin[x]')
plt.plot(x, cosy, c='r', linestyle='--', label='cos[x]')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.show()
```

**设置图标题**

```python
plt.plot(x, siny, label='sin[x]')
plt.plot(x, cosy, c='r', linestyle='--', label='cos[x]')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
plt.title('Weclome to the ML World!')
plt.show()
```

## 散点图

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color='red')
plt.show()
```

**二维正太分布散点图**

```python
x = np.random.normal(0, 1 , 10000)
y = np.random.normal(0, 1 , 10000)

plt.scatter(x, y, alpha=0.1) # alpha 设置点透明度
plt.show()
```
