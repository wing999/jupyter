### GroupBy方法

```python
group_salary = salary.groupby('工作岗位') # 根据岗位着一列，将相同公族哦岗位的 行 合并到一起
```

```python
group_salary.mean() # 平均值
```

```python
group_salary.sum() #求总和
```

```python
group_salary.describe() # 所有相关参数：数量 平均值 最大值....
```

```python
group_salary.describe().transpose() # 行和列反转
```
