# 线性回归及多项式回归分析

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

### 线性回归预测Pizza价格案例

```python
pizza_df= pd.read_csv('data/Pizza.csv')
pizza_df
```

```python
clf = LinearRegression()
clf.fit(pizza_df[['直径(英寸)']],pizza_df['价格(美元)'])
```

```python
clf.predict([[12]])
```

```python
plt.scatter(pizza_df[['直径(英寸)']],pizza_df['价格(美元)'],c='red')
x2=[[0],[12],[15],[25]]
y2=clf.predict(x2)
plt.plot(x2,y2)
```

### 线性回归预测波士顿房价

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
np.shape(boston.data)
# np.shape(boston.target)
```

```python
boston_df = pd.DataFrame(boston.data)
boston_df.columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
boston_df['MEDV']=boston.target
boston_df
```

```python
# 'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV'
data = boston_df[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
target = boston_df['MEDV']
x_train,x_test,y_train,y_test = train_test_split(data,target,random_state=0)
```
