# 特征选择测试

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
abalone = pd.read_csv('data/abalone.csv')
abalone['年龄']=abalone['环数']+1.5
abalone['性别'].replace(['M','F','I'],[0,1,2],inplace=True)
abalone
```

```python
# 通过性别与直径散点图

plt.scatter(abalone[abalone['性别']==0]['直径'],abalone[abalone['性别']==0]['环数'],c='red',label='M')
plt.scatter(abalone[abalone['性别']==1]['直径'],abalone[abalone['性别']==1]['环数'],c='green',label='F')
plt.scatter(abalone[abalone['性别']==2]['直径'],abalone[abalone['性别']==2]['环数'],c='blue',label='I')
plt.legend(loc=2) #左上角
plt.show()
```

```python
# '性别','长度','直径','高度','全重','去壳重量','内脏重量','壳体重量','环数','年龄'
data=abalone[['性别','长度','直径','高度','全重','去壳重量','内脏重量','壳体重量']]
target=abalone['环数']
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,target,random_state=0)
```

```python
poly = PolynomialFeatures(degree=5,include_bias=False)
poly.fit(X_train)
new_data = poly.transform(X_train)
```

```python
clf = LinearRegression()
clf.fit(new_data,y_train)
clf.score(new_data,y_train)
```

```python
poly_test = PolynomialFeatures(degree=5,include_bias=False)
poly_test.fit(X_test)
new_data_test = poly_test.transform(X_test)
clf.score(new_data_test,y_test)
```
